## Data Preperation

Data preperation will seek to remove unecesary data points that add no overall gain to the knowledege looking to eb understood. Within this stage missing values will be analysed and either an average or avaliable values within the same charatarstic will be applied or the intance will be removed. New fields will be created to denote important, identified characteristics that are not yet present within the dataset. Encoding of the data can also be appplied to ensure the outputs when read by a further stage of the M.L. pipeline the data is presented in an optimal layout.



```python
df = df.drop(columns=['instant', 'season', 'yr', 'mnth', 'holiday', 'workingday', 'dteday', 'atemp', 'casual', 'registered'])
df.head(1)
```

```python
df.isna()
```

```python
df['temp'] = df.groupby(['season','weathersit','hr'])['temp'].transform(lambda x: x.fillna(x.mean()))
df['atemp'] = df.groupby(['season','weathersit','hr'])['atemp'].transform(lambda x: x.fillna(x.mean()))
```
